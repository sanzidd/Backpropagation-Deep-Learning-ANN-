```python
import numpy as np 
import pandas as pd
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```python
df = pd.DataFrame([[8,8,1],[7,9,1],[6,10,0],[5,5,0]], columns=['cgpa', 'profile_score', 'placed'])
```

```python
df.shape
```

```python
X_train = df[['cgpa', 'profile_score']].values
y_train = df['placed'].values
```

```python
model = Sequential([
    # Input layer and first hidden layer with 2 units and sigmoid activation
    # input_shape=(2,) specifies 2 input features
    Dense(2, input_shape=(2,), activation='sigmoid'),
    # Output layer with 1 unit and sigmoid activation
    Dense(1, activation='sigmoid')
])
```

```python

model.compile(optimizer=SGD(learning_rate=0.0001), loss='binary_crossentropy')
```

```python
epochs = 50
history = model.fit(X_train, y_train, epochs=epochs, batch_size=1, verbose=0)
```

```python
for i, loss_val in enumerate(history.history['loss']):
    print(f"Epoch - {i+1} Loss - {loss_val:.10f}")
```

```python
print("\nKeras model summary:")
model.summary()
```

```python
print("\nFinal parameters (weights and biases) of the Keras model:")
for layer in model.layers:
    weights, biases = layer.get_weights()
    print(f"Layer: {layer.name}")
    print(f"  Weights:\n{weights}")
    print(f"  Biases:\n{biases}")
```
