```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]], columns=['cgpa', 'profile_score', 'lpa'])
```

```python
df.head()
```

```python
X_data = df[['cgpa', 'profile_score']].values
y_data = df['lpa'].values
```

```python
model = keras.Sequential([
    layers.Input(shape=(2,)), # Input layer with 2 features
    layers.Dense(2, activation='linear', name='hidden_layer'), # Hidden layer with 2 units
    layers.Dense(1, activation='linear', name='output_layer')  # Output layer with 1 unit
])
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss='mse')
```

```python
epochs = 300
history = model.fit(X_data, y_data, epochs=epochs, verbose=1)
```

```python
print("\nKeras Model Training Complete.")
print("Final Loss:", history.history['loss'][-1])
```

```python
w1, b1 = model.layers[0].get_weights()
print("\nKeras W1:\n", w1)
print("Keras b1:\n", b1)
```

```python
w2, b2 = model.layers[1].get_weights()
print("Keras W2:\n", w2)
print("Keras b2:\n", b2)
```
